# continue & Break

```python
course_subject  =  "python,java"

for alphabet in course_subject:
    if alphabet == ',':
        break # end the loop
    print(alphabet)
```

```python
course_subject  =  "python,java"

for alphabet in course_subject:
    if alphabet == ',':
        continue # skip that iteration 
        # iske baad wala code run nhi hoga 
    print(alphabet)
    
```

While Loops 

```python
# keeps running as long as condition is true '

# syntax 

# While conidition:
#     code to run
```

```python
count = 1 
# run until it is less 10 

while count <= 10 :
    print(count)
```

```python
# to end this loop, we have two things
count = 1

while count <= 10 :
    print(count)
    count += 1 # count = count + 1
    
```
