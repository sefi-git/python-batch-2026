```python
# try/except

try:
    # code that might cause an error
    age = int(input("Enter your age: "))
    print(age)  
except:
    # runs only if error happend in try:
    print("Not a valid Number!")

print("hahahahaha")
```

```python
# try/except

try:
    # code that might cause an error
    age = int(input("Enter your age: "))
    print(age)  
except ValueError:
    # runs only if error happend in try:
    print("Not a valid Number!")

print("hahahahaha")
```

```python
try:
    num = int(input("Enter a number: "))
    result = 100 / num
    print(result)
    
except ValueError:
    print("Not a Valid Number! ")

except ZeroDivisionError:
    print("Don't Enter 0: ")

```

```python
# raise 

try:
    a = int(input("Enter first number: "))
    if a == 0:
        print("yess")
        raise ZeroDivisionError("Error because of zero: ")
    result = 100/a
    print(result)
except Exception as e :
    print(f"Something went wrong! {e}")
```
