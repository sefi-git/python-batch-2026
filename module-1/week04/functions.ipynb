```python
# Write get_grade(score) that returns a letter grade.
# Then loop over a list of students and print each result.

def get_grade(score):
    if score >= 90:
        return "A"
    elif score >= 80:
        return "B"
    elif score >= 70:
        return "C"
    elif score >= 60:
        return "D"
    else:
        return "F"

students = [
    {"name": "Ali", "score": 92},
    {"name": "Sara", "score": 75},
    {"name": "Ahmed", "score": 55},
    {"name": "Zara", "score": 88}
]

for student in students:
    grade = get_grade(student["score"])
    print(f"{student['name']} got grade {grade}")
```

```python
# instead of creating multiple functions, we can do this in one and can return multiple values 

# multiple returns 
def get_stats(numbers):
    return max(numbers) , min(numbers), sum(numbers)


numbers = [5,2,4,7,9,10,12,2,3] # min max sum
max_no , min_n0, sum_no = get_stats(numbers)
print(max_no , min_n0, sum_no)
```

```python
count = 10 # global variable

def increment():
    count = 1 # local variable 
    count += 1 # count = count +1 # considering it as a local variable 
    print(count)
    return None

increment()
print(count)
```

```python
count = 10 # global variable

def increment():
    global count
    count += 1  
    print(count)

increment()
print(count)
```

```python
# class work 
# Build a receipt system using multiple functions that call each other.

# 1 function calculate_tax 
# 2 function calculate discount 
# 3 function caluculate total 
# 4th function print_receipt - tax, dicount payment

# 10000 + 1000 -2000 = 9000

```

```python
def calculate_tax(amount):
    return amount * 0.18

def calculate_discount(amount):
    return 2000

def calculate_total(subtotal, tax, discount):
    return subtotal + tax - discount

def print_receipt(subtotal):
    tax = calculate_tax(subtotal)
    discount = calculate_discount(subtotal)
    total = calculate_total(subtotal, tax, discount)

    print("Subtotal: ", subtotal)
    print("Tax:      ", tax)
    print("Discount: ", discount)
    print("Total:    ", total)

print_receipt(70000)





```
