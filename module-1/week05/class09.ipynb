loop inside another loop

```python
for i in range(3): # 3 times 0,1,2
    for j in range(3): # 3 times 0,1,2
        print(i,j)
        
    
# first iteration i = 0 (j = 0,1,2)
# second iteration i = 1 (j = 0,1,2)
# thirdt iteration i = 2 (j = 0,1,2)
```

Multiplication Table

```python
for i in range(2,4): # 1, 2, 3
    for j in range(1,11):
        print(f'{i} X {j} = {i*j}')
    print()
```

```python
# * 
# **
# ***
# ****
# *****

for i in range(1,6):# 1,2,3,4,5
    for j in range(i): 
        print('*',end=" ")
    print()

    if i == 5:
        break

    # first iterations i = 1 (range(1)) ---- j loop executed once
    # second iteration i = 2 (range(2)) ---- j loop executed twice
    # third iteration i = 3  (range(3)) ---- j loop executed thrice
    # 
    # fift iteration i = 5 (range(5))  ----- j loop executed five times
```

```python
# * * * * *
# * * * *
# * * *
# * *
# *


for i in range(1,6):
    for j in range(6-i):
        print("*",end=" ")
    print()

    # first iteration i = 1 (j : 6 -1 = 5)
    # second iteation i = 2 (j = 6- 2 = 4)
```

Matrix -- Nested List

```python
matrix = [
    [1,2,3], # 0
    [4,5,6], # 1
    [7,8,9]  # 2
]

# row horizontal line 
# column vertical line  3 X 3

matrix[1] # [4,5,6] 0,1,2
matrix[1][2]
matrix[2][2]
matrix[2][-1]
matrix[1][-1]

```

```python
# print matrix values with nested loop 

matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]  # 2
]

for row in matrix:
    for item in row:
        print(item, end=" ")
    print()
```

```python
groceries = [
    ['apple','mangoes','dates'],
    ["chilies","lemon","ladyfinger"]
]

# index 0 -- ['apple','mangoes','dates']
# index 1 -- ["chilies","lemon","ladyfinger"]

groceries[0][1]
groceries[1][2]

for row in groceries:
    for item in row:
        print(item)
```
