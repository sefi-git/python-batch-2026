```python
# Arithmetic Operators

a = int(input("Enter first value: ")) # 10
b = int(input("Enter Second value: ")) # 3

print(a + b)    # 13  — addition
print(a - b)    # 7   — subtraction
print(a * b)    # 30  — multiplication
print(a / b)    # 3.333...  — division (always float)
print(a // b)   # 3   — floor division (whole number only)
print(a % b)    # 1   — modulo (remainder)
# 10/3 ...3,6,9   
# 10 - 9 = 1 
# 15/4 --    15 - 12  =3
print(a ** b)   # 1000 — power (10 to the power of 3)
# 2 * 2  =4 
# 2 *2 *2 =6
# 10 * 10 *10 = 1000
```

```python
# Logical Operators
# Combine multiple comparisons together.
# AND OR

age  = 20
score = 85

print(age >= 18 and score >= 80)    # True  — both must be True
print(age >= 18 and score >= 90)    # False — score condition fails
print(age < 18 or score >= 80)     # True  — at least one is True
print(not True)                     # False — flips True to False
print(not False)                    # True  — flips False to True
```

```python
# Exercise 3 — Age in Days
# Ask the user for their age in years. Calculate and print their approximate age in days, hours, and minutes.

age_years   = int(input("Enter your age in years: "))

age_days    = age_years * 365
age_hours   = age_days * 24
age_minutes = age_hours * 60

print(f"You are approximately:")
print(f"  {age_days} days old")
print(f"  {age_hours} hours old")
print(f"  {age_minutes} minutes old")
```
