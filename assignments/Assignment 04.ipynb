## **Q3 — Loop Through Dictionary**
* person = {"name": "Ali","age": 25,"city": "Karachi"}
* Print keys and values using loop.

```python
person = {'name':'Ali','age':25,'city':'Karachi'}
for key in person.keys():
    print (key)
for value in person.values():
    print(value)
```

## **Q4 — Count Frequency of Numbers**
* numbers = [1,2,2,3,3,3,4,4,4,4]
* Create frequency dictionary.

```python
numbers=[1,2,2,3,3,3,4,4,4,4]
frequency = {
'frequency of 1': numbers.count(1),
'frequency of 2': numbers.count(2),
'frequency of 3': numbers.count(3),
'frequency of 4': numbers.count(4),
}
print(frequency)
```

## **Q5 — Word Counter** 
* Take sentence input and count word frequency.

```python
sentence=input("Enter sentence: ")
print(sentence)
sentence=sentence.replace(" ","")
sentence=sentence.lower()
for word in sentence:
    print("Frequency of",word,':',sentence.count(word))
```

## **Q6 — Student Marks & Average** 
* marks = {"Math":80, "English":70, "Physics":90}
* Calculate total and average.

```python
marks = {
    'Math':80,
    'English':70,
    'Physics':90
}
total=0
for value in marks.values():
    total+=value
    avg=total/3
print('Total:',total)
print('Average',avg)
```

## **Q7 — Highest Scoring Student** 
* students = {"Ali":85,"Sara":92,"Ahmed":78,"Zara":95}
* Find highest scoring student

```python
student={"Ali":85,"Sara":92,"AHmed":78,"Zara":95}
max_value=max(student.values())
for key,value in student.items():
    if value==max_value:
        print(f"The highest scoring student is: {key}")
```
