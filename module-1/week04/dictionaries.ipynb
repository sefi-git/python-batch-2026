```python
# Question No 02 

students ={
    "Ali":[80,90,70],
    "Ayan":[100,50,20],
    "Aftab":[60,57,66]
}

# calculate and print the avg marks of every student 

for name, marks in students.items():
    avg = sum(marks) / len(marks)
    print(name, "average marks =", avg)
```

```python
products ={
    "apple":80,
    "mango":200,
    "banana": 50,
    "oranges":180
}

expensive_products = {} # price more than 100

max_price = 100
for fruit,price in products.items():
    if price > max_price:
        expensive_products[fruit] = price
print(expensive_products)
```

Tuples

```python
# same like a list, stores multiple items 
# difference: immutable (can't be changed once created)
# syntax starts with round brackets 

# list 
fruits = ["apple","mango","grapes"]
fruits[0] = 'dates'

# tuple
fruits = ("apple","mango","grapes")
fruits[0] = 'dates'
```
