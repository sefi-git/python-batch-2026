# with 

```python
# with -> opens file and automatically close it when the block ends 

with open("notes.txt","r") as file_read:
    content = file_read.read()
    print(content)
```

```python
content = file_read.read()
```

```python
# Class Acitivity 

# Write a program that:

# Asks the user to enter a message
# Appends it to log.txt with the entry number
# Keeps asking until user types "quit"
# At the end, reads and prints the full log

```

```python
# Write a program that:

user_input = 1

file = open("log.txt", "a")

while True:

    user_text = input("Enter text: ")

    if user_text.lower() == "quit":
        break

    file.write(f"{user_input}: {user_text}\n")
    user_input += 1

   
# open it and read it
print(file.read())

file.close()





```
