3. Write a program in which use all the operators we can use in Python

```python
# Write your code here
#op1="+"
#op2="-"
#op3="*"
#op4="/"
#op5="%"
#op6="//"
num1=10
num2=2
print("The use of all the operators is demonstrated as:")
ans1=num1+num2
print(f"'+' operator: {num1} + {num2} = {ans1}")
ans2=num1-num2
print(f"'-' operator: {num1} - {num2} = {ans2}")
ans3=num1*num2
print(f"'*' operator: {num1} * {num2} = {ans3}")
ans4=num1/num2
print(f"'/' operator: {num1} / {num2} = {ans4}")
ans5=num1%num2
print(f"'%' operator: {num1} % {num2} = {ans5}")
ans6=num1//num2
print(f"'//' operator: {num1} // {num2} = {ans6}")
```
